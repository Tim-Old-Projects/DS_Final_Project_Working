```python
#count missing values
df2['Full FIPS'].isna().sum()
```

```python
#remove missing values
df2_no_missing = df2['Full FIPS'].dropna()
```

```python
#Formatting change to integer
fips = df2_no_missing.astype(dtype = int)
```

```python
#Currently has 3 extra digits
fips.sort_values(ascending=True)
```

```python
#Convert to string and remove unneeded digits
string_fips = fips.astype(dtype = str)
string_fips_sliced = string_fips.str.slice(0,-3)
```

```python
#convert back to dataframe
string_fips_sliced_df = pd.DataFrame(string_fips_sliced)
```

```python
#Add target variable
string_fips_sliced_df['target_2'] = 1
```

```python
#Check dataframe
string_fips_sliced_df.head()
```

```python
#Reassign integer type to Full FIPS
string_fips_sliced_df['Full FIPS'] = string_fips_sliced_df['Full FIPS'].astype(dtype = int)
```

```python
#Merge Target Variable with Census Data
cleaned_df = pd.merge(df, string_fips_sliced_df, left_on='GIDBG' , right_on='Full FIPS', how='left', )
```

```python
#Clean target variable, replace NA with 0
cleaned_df['target_2'] = cleaned_df['target_2'].fillna(0)
```

```python
#Make target variable an integer type
cleaned_df['target_2'] = cleaned_df['target_2'].astype(dtype = int)
```
